```python
#Step:3 :Data visualization
import matplotlib.pyplot as plt
#plt.scatter(x-axia, y-axis)
plt.scatter(df['YearsExperience'], df['Salary'])
plt.title('Salary data')
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
```

```python
#Step: 4 Divide the data into input(x) and output(y)
#Input(x) - is always 2 dimensional array - YearsExperience column
#Output(y) - is 1 dimensional- Salary column
#In the column slicing part, if there is a ':' then array is two dimensional
x = df.iloc[0:30,0:1].values
x #.values convert data frame into an array
```

```python
y = df.iloc[0:30,1].values
y
```

```python
#Step5 : Not neccessary
```

```python
#Step: 6 Not neccessary
```

```python
#Step: 7
#Step: 7 Apply a Regressor/Classifier or Cluster
#sklearn.linaer_model- Package, LinearRegression is a lib
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
#Step:8 Fit the model- Mapping/Ploting the i/p values with o/p
model.fit(x, y) #linearRegression.fit(x, y)
#we hv plotted x And y values LinearRegression library
```

```python
#Step: 9 Predict the output
y_pred = model.predict(x) #using the i/p values we predict the o/p
y_pred  #Predicted o/p values
```

```python
y
```

```python
#Individual prediction
model.predict([[7.5]])
```

```python
#Visualization for the best fit line
plt.scatter(x, y)
plt.plot(x, y_pred, c='Orange')
plt.title('BEST fit LINE')
plt.xlabel('YearsExp ')
plt.ylabel('Salary')
```
